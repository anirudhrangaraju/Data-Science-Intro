```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer,CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import pickle
```

```python
data=pd.read_csv("C:\\Users\\A\\Desktop\\spooky_identification\\test.csv")
```

```python
X_test = data.text
```

```python
#Loading Naive Bayes Model
```

```python
with open ("model_nb.pkl","rb") as f:
    model_nb=pickle.load(f)
```

```python
#Loading SVM Model
```

```python
with open ("model_svm.pkl","rb") as f:
    model_svm=pickle.load(f)
```

```python
#Loading Decision Tree Model
```

```python
with open ("model_dt.pkl","rb") as f:
    model_dt=pickle.load(f)
```

```python
#Loading Random Forest Model
```

```python
with open ("model_rf.pkl","rb") as f:
    model_rf=pickle.load(f)
```

```python
#Prediction of Various Models
```

```python
def nb():
    
    predicted_nb_score = model_nb.predict(data.text)
    predicted_nb_score

    predicted_output = pd.DataFrame(predicted_nb_score)
    
    submission_file = pd.concat([predicted_output,X_test],axis=1)
    submission_file.rename(columns={"text":"Author_Book"},inplace=True)
    submission_file.rename(columns={0:"Author_Code"},inplace=True)
    
    submission_file.to_csv("predicted_output_nb.csv")
    return submission_file
    
```

```python
nb()
```
